# Plotting
Using matplotlib

```python
import matplotlib.pyplot as plt
```

```python
plt.figure
plt.plot(data[:,1])
```

```python
plt.figure
samplename = headers[1]
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (angstroms)')
plt.plot(data[:,1], label=samplename)
plt.legend()
plt.savefig(f'{samplename}.png', dpi=300)
```

```python
plt.figure
samplename = headers[1]
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (angstroms)')
plt.plot(data[:,1], label=samplename, color='#e1c2f2')
plt.plot(data[:,2], label=headers[2], color='#bdf0e7')
plt.legend()
plt.savefig(f'twosamples.png', dpi=300)
```

```python
plt.figure
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (angstroms)')
for col in range(1, num_columns):
    plt.plot(data[:, col], label=headers[col])
plt.legend()
plt.savefig('allsamples.png')
```

```python
datamin = np.min(data[:, 1:])
datamax = np.max(data[:, 1:])

for col in range(1, num_columns):
    samplenames = headers[col]
    plt.figure()
    plt.ylim(datamin, datamax)
    plt.xlabel('Simulation Frame')
    plt.ylabel('Distance (angstroms)')
    plt.plot(data[:, col], label=samplenames)
    plt.legend()
    plt.savefig(f'{samplenames}.png')
```
